```python
import matplotlib.pyplot as plt
df.boxplot(column='age')
plt.title("age distribution boxplot")
plt.show()
```

```python
#passenger class
sns.countplot(x='class',data=df)
plt.title("passenger class distribution")
plt.show()
```
